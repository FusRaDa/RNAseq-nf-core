### Step 1: Initial run with BAM generation


```python
# Set nextflow work environment
%env NXF_WORK=nf_temp/work
```

```python
# Ensure system has at least 12 cores
import multiprocessing
cores = multiprocessing.cpu_count()
print(f"Number of CPU cores available: {cores}")
```

```python
import os
import csv
```

```python
if cores >= 12:
  with open('/content/drive/MyDrive/Colab_Notebooks/RNA-seq-nf/genome_refs.csv', mode='r', newline='', encoding='utf8') as file:
      csv_reader = csv.reader(file, delimiter=',')
      for row in csv_reader:
          fasta_file_path = row[0]
          folder_name = os.path.basename(fasta_file_path)

          # Run nf-core/rna-seq pipeline
          !nextflow run nf-core/rnaseq \
          --input /content/drive/MyDrive/Colab_Notebooks/RNA-seq-nf/single-end.csv \
          --outdir nf_temp/$folder_name \
          --gtf /content/drive/MyDrive/Colab_Notebooks/RNA-seq-nf/Homo_sapiens.GRCh38.115.gtf.gz \
          --fasta $fasta_file_path \
          -profile conda

          # zip results file
          !zip -r /content/nf_temp/$folder_name.zip /content/nf_temp/$folder_name

          # download to local machine
          from google.colab import files
          files.download(f"/content/nf_temp/{folder_name}.zip")
else:
  print("Switch environment to at least 12 cores.")
```
